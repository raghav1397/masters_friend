```python
import re
import time
import random
import requests
import pandas as pd
from lxml import html
from multiprocessing.pool import Pool
from selenium.webdriver.common.by import By
from selenium.webdriver.common.proxy import Proxy, ProxyType
from selenium.webdriver import Chrome, DesiredCapabilities, Remote, Firefox
```

```python
with open("proxies.txt", "r") as f:
    proxies = f.readlines()
proxies = [p.split('\n')[0] for p in proxies]
```

```python
def check_proxy(p):
    driver = Chrome("./chromedriver")
    driver.get('https://scamalytics.com/ip/'+p.split(':')[0])
    score = driver.find_element_by_class_name('score')
    score = int(score.text.split()[-1])
    driver.close()
    if score < 25:
        return p
    else:
        return '0'
```

```python
pool = Pool(12)
checked_proxies = pool.map(check_proxy, proxies)
working_proxies = [c for c in checked_proxies if c!='0']
```

```python
with open('working_proxies.txt', 'w') as f:
    f.writelines('\n'.join(working_proxies))
```
